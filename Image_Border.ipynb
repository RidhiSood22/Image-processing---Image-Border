```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/download.jpeg')
```

```python
cv2_imshow(image)
```

```python
image1 = cv2.copyMakeBorder(image, 10, 10, 10, 10, cv2.BORDER_CONSTANT, value=(225, 225, 225))
```

```python
cv2_imshow(image1)
```

```python
image2 = cv2.copyMakeBorder(image,100,100,50,50,cv2.BORDER_REFLECT)
```

```python
cv2_imshow(image2)
```
